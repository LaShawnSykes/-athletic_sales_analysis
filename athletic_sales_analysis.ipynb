```python
# Import Libraries and Dependencies
import pandas as pd
```

### 1. Combine and Clean the Data
#### Import CSVs

```python
# Read the CSV files into DataFrames.
df1 = pd.read_csv('athletic_sales_2020.csv')
df2 = pd.read_csv ('athletic_sales_2021.csv')
```

```python
# Display the 2020 sales DataFrame
from IPython.display import display, HTML
display(df1)
```

```python
# Display the 2021 sales DataFrame
display(df2)
```

#### Check the data types of each DataFrame

```python
# Check the 2020 sales data types
df1.dtypes
```

```python
# Check the 2021 sales data types.
df2.dtypes
```

#### Combine the sales data by rows.

```python
# Combine the 2020 and 2021 sales DataFrames on the rows and reset the index
df3 = pd.concat([df1,df2], ignore_index=True)
display(df3)
```

```python
# Check if any values are null.
nv = df3.isnull().sum()
display(nv)
print(nv)
```

```python
# Check the data type of each column
df3.dtypes
```

```python
#Compare df1,df2,df3 data types
cr = (df1.dtypes == df2.dtypes) & (df2.dtypes == df3.dtypes)
display(cr)
```

```python
# Convert the "invoice_date" to a datetime datatype
df3['invoice_date'] = pd.to_datetime(df3['invoice_date'], format ='%m/%d/%y')
unique_region = df3['region'].unique()
display(unique_region)
display(df3)
```

```python
# Confirm that the "invoice_date" data type has been changed.
df3.dtypes
```

### 2. Determine which Region Sold the Most Products

#### Using `groupby`

```python
# Get unique values for region, state, and city
unique_region = df3['region'].unique()
unique_state = df3['state'].unique()
unique_city = df3['city'].unique()

print("Unique regions:", unique_region)
print("Unique states:", unique_state)
print("Unique cities:", unique_city)
print("----------------------------------------------------------------------------------")

# Show the number of products sold for region, state, and city
products_sold = df3.groupby(["region", "state", "city"]).size().reset_index(name='Total_Products_Sold')

# Show the top 5 results
top_5_results = products_sold.nlargest(5, "Total_Products_Sold")
print("Top five regions, states, and cities with the most products sold:")
print(top_5_results)
print("-----------------------------------------------------------------------------------------------")

# Find the top region among the top 5
top_region = top_5_results.head(1)
print("\033[1mTop region that sold the most products:")
print(top_region)          
```

#### Using `pivot_table`

```python
unique_region = df3['region'].unique()
unique_state = df3['state'].unique()
unique_city = df3['city'].unique
display(unique_region,unique_state,unique_city)
print("-----------------------------------------------------------------------------------------------")

# Show the number products sold for region, state, and city.
product_sold_pt = df3.pivot_table(index=["region", "state","city"],values='units_sold', aggfunc='count')
display(product_sold_pt.head(15))

# Rename the sum to "Total_Products_Sold".
product_sold_pt = product_sold_pt.rename(columns ={'units_sold':'Total_Products_Sold'})
display(product_sold_pt.head(15))

# Show the top 5 results.
top_5_results = product_sold_pt.nlargest(5,"Total_Products_Sold") 
display(top_5_results) 
print("-----------------------------------------------------------------------------------------------")

# Find the top region among the top 5
top_region = top_5_results.nlargest(1, "Total_Products_Sold")
display(HTML("<b>Top region that sold the most products:</b>"))
display(top_region)
```

### 3. Determine which Region had the Most Sales

#### Using `groupby`

```python
df3.info()
print('-----------------------------------------------------------------------')

# Show the total sales for the products sold for each region, state, and city.
total_sales = df3.groupby(['region','state','city'])['total_sales'].sum().reset_index()

# Rename the "total_sales" column to "Total Sales"
total_sales = total_sales.rename(columns ={'total_sales':'Total_Sales'})
total_sales['Total_Sales'] = total_sales['Total_Sales'].apply(lambda x: "{:,}".format(x))                         
display(total_sales.head(15))

# Convert the "Total_Sales" column to a numeric data type
total_sales['Total_Sales'] = total_sales['Total_Sales'].str.replace(',', '').astype(int)

# Show the top 5 results.
top_5_results = total_sales.nlargest(5,"Total_Sales") 
top_5_results['Total_Sales'] = top_5_results['Total_Sales'].apply(lambda x: "{:,}".format(x))
display(top_5_results.head(15))                         
```

#### Using `pivot_table`

```python
df3.info()
print('-----------------------------------------------------------------------')

# Show the total sales for the products sold for each region, state, and city.
pivot_total_sales = df3.pivot_table(index=['region', 'state', 'city'], values='total_sales', aggfunc='sum').reset_index()

# Rename the "total_sales" column to "Total Sales"
pivot_total_sales = pivot_total_sales.rename(columns={'total_sales': 'Total Sales'})

# Convert the "Total Sales" column to a numeric data type
pivot_total_sales['Total Sales'] = pivot_total_sales['Total Sales'].astype(int)

# Show the top 5 results
top_5_results = pivot_total_sales.nlargest(5, "Total Sales")
top_5_results['Total Sales'] = top_5_results['Total Sales'].apply(lambda x: "{:,}".format(x))

# Display the top 5 results
display(top_5_results.head(5))
```

### 5. Determine which Retailer Sold the Most Women's Athletic Footwear

```python
# Filter the sales data to get the women's athletic footwear sales data.
women_athletic_footwear_sales = df3[df3['product'] == "Women's Athletic Footwear"]

# Group the data by retailer and sum the sales
total_sales_by_retailer = women_athletic_footwear_sales.groupby('retailer')['total_sales'].sum()

# Find the retailer that sold the most women's athletic footwear
top_retailer = total_sales_by_retailer.idxmax()

# Display the retailer that sold the most women's athletic footwear
print(f"The retailer that sold the most women's athletic footwear is: {top_retailer}")
```

#### Using `groupby`

```python
# Show the total number of women's athletic footwear sold for each retailer, region, state, and city.
unique_products = df3['product'].unique()
print(unique_products)
print('----------------------------------------------------------------------------------------------------------------------------------')

# Filter the sales data to get the women's athletic footwear sales data
women_athletic_footwear_sales = df3[df3['product'] == "Women's Athletic Footwear"]
# Group the data by retailer, region, state, and city and sum the units sold
women_footwear_units_sold = women_athletic_footwear_sales.groupby(['retailer', 'region', 'state', 'city'])['units_sold'].sum().reset_index()

# Rename the "units_sold" column to "Womens_Footwear_Units_Sold"
women_footwear_units_sold = women_footwear_units_sold.rename(columns={'units_sold': 'Womens_Footwear_Units_Sold'})

# Show the top 5 results
top_5_results = women_footwear_units_sold.nlargest(5, "Womens_Footwear_Units_Sold")

# Display the top 5 results
print(top_5_results)
```

#### Using `pivot_table`

```python
# Show the total number of women's athletic footwear sold for each retailer, region, state, and city.
unique_products = df3['product'].unique()
print(unique_products)
print('----------------------------------------------------------------------------------------------------------------------------------')


# Filter the sales data to get the women's athletic footwear sales data
women_footwear_pivot = pd.pivot_table(df3[df3['product'] == "Women's Athletic Footwear"], values='units_sold', index=['retailer', 'region', 'state', 'city'], aggfunc='sum').reset_index()

# Rename the "units_sold" column to "Womens_Footwear_Units_Sold"
women_footwear_pivot = women_footwear_pivot.rename(columns={'units_sold':'Womens_Footwear_Units_Sold'})

# Show the top 5 results.
top_5_results = women_footwear_pivot.nlargest(5, "Womens_Footwear_Units_Sold")

# Displaying the top 5 results
display(top_5_results)
```

### 5. Determine the Day with the Most Women's Athletic Footwear Sales

```python
# Create a pivot table with the 'invoice_date' column is the index, and the "total_sales" as the values.
print(df3.columns)
print('----------------------------------------------------------------------------------------------------------------------------------')
pt_invoice_date = pd.pivot_table(df3,index = 'invoice_date', values = 'total_sales')

# Optional: Rename the "total_sales" column to "Total Sales"
pt_invoice_date = pt_invoice_date.rename(columns={'total_sales': 'Total Sales'})


# Reformat "Total Sales" column with commas and to the second decimal place
pt_invoice_date['Total Sales'] = pt_invoice_date['Total Sales'].apply(lambda x: "{:,.2f}".format(x))

# Find the day with the most women's athletic footwear sales
day_highest_sales = pt_invoice_date['Total Sales'].idxmax()

# Show the table
print(f"The day that sold the most women's athletic footwear is: {day_highest_sales}")
```

```python
print(df3.dtypes)
print("--------------------------------------------------------------------------")

# Resample the pivot table into daily bins and get the total sales for each day
resampled_pt = pt_invoice_date.resample('D').sum()

# Convert "Total Sales" column to numeric data type
resampled_pt['Total Sales'] = resampled_pt['Total Sales'].str.replace(',', '').astype(float)

# Sort the resampled pivot table in ascending order on "Total Sales"
sorted_resampled_pt = resampled_pt.sort_values(by='Total Sales', ascending=True)


# Displaying the sorted, resampled pivot table
display('The sorted, resampled pivot table is :')
display(sorted_resampled_pt)
```

### 6.  Determine the Week with the Most Women's Athletic Footwear Sales

```python
print(df3.dtypes)
print("--------------------------------------------------------------------------")

# Filter the data to include only women's athletic footwear sales
women_athletic_footwear_sales = df3[(df3['product'] == "Women's Athletic Footwear")]

# Group the data by week and sum the sales
weekly_women_athletic_sales = women_athletic_footwear_sales.resample('W', on='invoice_date')['total_sales'].sum()

# Find the week with the highest sales
week_with_highest_sales = weekly_women_athletic_sales.idxmax()

# Find the top five weeks with the highest sales
top_five_weeks = weekly_women_athletic_sales.nlargest(5)
display("Top five weeks with the highest women's athletic footwear sales:")
display(top_five_weeks)
print("--------------------------------------------------------------------------")


# Display the week with the most women's athletic footwear sales
display("Week with the most women's athletic footwear sales:", week_with_highest_sales)
```
